```python
# Question 1 - Age Categorization

def get_age_category(age):
  if age < 3:
    return "Under 3"
  elif age >= 3 and age <= 12:
    return "Ages 3-12"
  else:
    return "Over 12"

# Example usage:
customer_ages = [1, 2, 5, 10, 12, 15, 20, 0]
print("Age Binning Results:")

# Initialize age_bins to count customers in each category
# Index 0: Under 3, Index 1: Ages 3-12, Index 2: Over 12
age_bins = [0, 0, 0]

for age in customer_ages:
    category = get_age_category(age)
    print(f"Age: {age}, Category: {category}")

    # Increment the appropriate bin.
    # Each customer's age is processed exactly once by the `for` loop.
    # The `get_age_category` function ensures that each age falls into one and only one category.
    # Therefore, each customer will increment only one counter in `age_bins` exactly once.
    if category == "Under 3":
        age_bins[0] += 1
    elif category == "Ages 3-12":
        age_bins[1] += 1
    else: # Over 12
        age_bins[2] += 1

print("\nCategory Counts:")
print(f"Under 3: {age_bins[0]} customers")
print(f"Ages 3-12: {age_bins[1]} customers")
print(f"Over 12: {age_bins[2]} customers")

# Question 2 - Sorted by Age
# This section implements a selection sort-like algorithm to sort customer groups by age.
# The criteria is to repeatedly identify the group with the smallest age, move it to a new list,
# and remove it from the original list.

sorted_groups = []

# Initialize group_list with example data for sorting.
# Each inner list represents a 'group' and is assumed to have an age at index 2 for comparison.
# For example: [group_name, other_data, age]
# If this list is empty, the sorting logic will not execute.
group_list = [['Group A', 'details1', 15], ['Group B', 'details2', 5], ['Group C', 'details3', 20], ['Group D', 'details4', 2]]

# The while loop continues as long as there are elements left in the original group_list.
while group_list:
    # Assume the first element in the current group_list is the smallest.
    min_group = group_list[0]

    # Iterate through all groups in the remaining group_list to find the actual minimum.
    for group in group_list:
        # The condition `group[2] < min_group[2]` compares the age (at index 2) of the current group
        # with the age of the currently identified minimum group.
        if group[2] < min_group[2]:
            min_group = group # If a smaller age is found, update min_group.

    # Once the smallest group in the current iteration is found, add it to the sorted list.
    sorted_groups.append(min_group)
    # Then, remove it from the original group_list to avoid re-selecting it and to eventually
    # terminate the while loop.
    group_list.remove(min_group)

print("\nSorted Groups by Age (youngest to oldest):")
for group in sorted_groups:
    print(f"Group: {group[0]}, Age: {group[2]}")

# Question 3 - Total Revenue

def total_gross_revenue(customers):
  CHILD_PRICE = 8
  ADULT_PRICE = 12

  total = 0

  for name, age in customers:
      # Booleans evaluate to 1 (True) or 0 (False)
      is_child = (age >= 3) and (age <= 12)
      is_adult = (age > 12)

      total += (is_child * CHILD_PRICE) + (is_adult * ADULT_PRICE)

  return total

# Question 4 - Averages across Groups

def group_averages(groups):
    # stores the sum of all the customers ages
    total_age = 0
    # stores the sum of all the coupon values
    total_coupon = 0
    # counts the total number of customer between all the groups
    total_customers = 0

    # Loops each customer in their list group
    for group in groups:
      # Loops each customer in their current group
        for age, coupon in group:
          # adds the customers age to the total
            total_age += age
            # adds the customers coupon value to the total
            total_coupon += coupon
            # increases the customer count +1
            total_customers += 1
    # calculates the average age by dividing age by # of customers
    avg_age = total_age / total_customers
    # calculates the average group size using total customers and number of groups
    avg_group_size = total_customers / len(groups)
    # calculates the average coupon value by dividing the total coupons by total customers
    avg_coupon = total_coupon / total_customers
    # returns the three averages as a list
    return [avg_age, avg_group_size, avg_coupon]
```
